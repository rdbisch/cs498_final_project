```python
import pandas as pd
import glob
import datetime
```

```python
files = glob.glob("../data/*.csv")
```

```python
result = pd.DataFrame()
for f in files:
    terms = f.split("_")
    print(terms)
    if terms[0] == "../data/smart": continue
    device_no = int(terms[1])
    terms2 = terms[2].split('.')
    date_time = datetime.datetime.strptime(terms2[0], "%Y-%m-%dT%H")
    
    temp = pd.read_csv(f)
    temp["device"] = device_no
    temp["file"] = date_time
    result = result.append(temp)
```

```python
result_dedup = result.drop_duplicates(subset=['datetime', 'device'])
```
