```python
import pandas as pd
```

```python
import datetime
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
def cleanDate(s):
    return datetime.datetime.strptime(s, "%Y-%m-%d %H:%M:%S.%f")
```

```python
all_devices = pd.read_csv("consolidated.csv")
all_devices["time"] = all_devices["datetime"].apply(cleanDate)
```

```python
all_devices["time"]
```

```python
fig, ax = plt.subplots(figsize=(12,12))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d %H:%M'))
fig.autofmt_xdate()

all_devices2 = all_devices[all_devices["temperature"] < 60]
all_devices2 = all_devices2[all_devices2["temperature"] > 5]
all_devices2 = all_devices2[all_devices2["time"] > datetime.datetime(2021, 5, 7)]
all_devices2 = all_devices2[all_devices2["time"] < datetime.datetime(2021, 5, 9)]


sns_plot = sns.scatterplot(data=all_devices2,
                x="time",
                y="temperature", 
                hue="device",
                size=1,
                linewidth=0,
                palette="tab10",
                alpha=0.3,
                ax=ax)

fig.savefig("no_corr.png")
```
